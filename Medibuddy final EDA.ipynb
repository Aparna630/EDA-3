Q1. Does the gender of the person matter for the company as a constraint for extending policies?

````python
# Chart - 1 visualization code

df.groupby('sex')['charges in INR'].mean()
````

````python
plt.figure(figsize=(6,4))
sns.violinplot(
    x='sex',
    y='charges in INR',
    data=df,
    inner='quartile'
)
plt.title("Distribution of Insurance Charges by Gender")
plt.show()
````

The violin plot illustrates the distribution of insurance claim amounts across male and female policyholders. The width of each violin represents the density of data points at different charge levels, while the internal quartile lines indicate the median and interquartile range.

From the visualization, both male and female distributions show significant overlap, with similar medians and comparable spread of claim amounts. The majority of claims for both genders lie within the same cost range, and high-value claims are observed for both groups. This indicates that the pattern and variability of insurance charges are nearly identical across genders.




````python
gender_charges = df.groupby('sex')['charges in INR'].sum()

plt.figure(figsize=(6,6))
plt.pie(
    gender_charges,
    labels=gender_charges.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Share of Total Insurance Charges by Gender")
plt.show()
````

The pie chart represents the proportion of total insurance charges contributed by male and female policyholders. It shows that male policyholders account for approximately 53% of the total insurance expenditure, while female policyholders contribute around 47%.

The marginal difference in contribution suggests that overall insurance spending is almost evenly distributed between genders. No single gender disproportionately impacts the company’s total insurance cost.

Q8. Can the company extend certain discounts after checking the health status (BMI) in this case?

````python
def bmi_category(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif bmi < 25:
        return 'Normal'
    elif bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'

df['BMI_Category'] = df['bmi'].apply(bmi_category)
````

````python
df.groupby('BMI_Category')['charges in INR'].mean()
````

````python
plt.figure(figsize=(7,4))
sns.barplot(
    x='BMI_Category',
    y='charges in INR',
    data=df,
    ci=None
)
plt.title("Average Insurance Charges by BMI Category")
plt.xlabel("BMI Category")
plt.ylabel("Average Charges in INR")
plt.show()
````

Q8. Health-Based Discounts Using BMI
To evaluate whether discounts can be offered based on health status, policyholders were classified into BMI categories such as underweight, normal, overweight, and obese. The average insurance claim amount was analyzed for each category.

The results show that individuals with a normal BMI incur the lowest average insurance charges, while overweight and obese individuals have significantly higher claim amounts. This indicates a clear relationship between BMI category and insurance expenditure.

Inference:
The company can extend discounts or incentives to policyholders with healthy BMI levels, as they pose lower insurance risk. Health-based pricing can encourage healthier lifestyles while helping reduce overall claim costs.
